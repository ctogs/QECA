```python
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True
```

```python
last_numbers = [39, 30, 48, 28, 42, 48, 34, 46, 38, 67, 88, 40, 98, 78, 92, 59, 53, 3, 91, 39]
def valid_proof(last_numbers):
    backend = Aer.get_backend('qasm_simulator')
    quantum_instance = QuantumInstance(backend=backend, shots=1000)
    prime_nums = []
    
    for n in last_numbers:
        if is_prime(n):
            continue
        for a in range(n):
            try:
                print(n, a)
                my_shor = Shor(N=n, a=a, quantum_instance=quantum_instance)
                result = Shor.run(my_shor)
                z = result.get('factors')[0]
                print(z)
                prime_nums.append(z)
                break
            except:
                continue

    return prime_nums
```

```python
valid_proof(last_numbers)
```
